```python
#imports
import pandas as pd
import numpy as np
```

```python
#load the dataset into pandas dataframe datasets from kaggel movie recommendation dataset
df = pd.read_csv('../data/movies_metadata.csv')
df.head()
#reads the first 80% by vote_count desc and assign it to the variable m
m = df['vote_count'].quantile(0.80)
m
```

```python
#Only consider movies longer than 45 minutes and shorter than 300 minutes
q_movies = df[(df['runtime'] >= 45) & (df['runtime'] <= 300)]

#Only consider movies that have garnered more than m votes
q_movies = q_movies[q_movies['vote_count'] >= m]

#Inspect the number of movies that made the cut
q_movies.shape
```

```python
# Calculate C
C = df['vote_average'].mean()
C
```

```python
# Function to compute the IMDB weighted rating for each movie dont get confused just using IMDB's weight and sum(it is gonna help as in filtering out recommendations not only based on rating and also vote count also other factors that is what the func retutns)
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    # Compute the weighted score
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
# Compute the score using the weighted_rating function defined above
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
#Sort movies in descending order of their scores
q_movies = q_movies.sort_values('score', ascending=False)

#Print the top 25 movies
q_movies[['title', 'vote_count', 'vote_average', 'score', 'runtime']].head(25)
```
